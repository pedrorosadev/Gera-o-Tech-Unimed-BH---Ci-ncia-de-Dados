```python
#Criando nosso DataFrame
df = pd.read_excel('AdventureWorks.xlsx')
```

```python
#Visualizando as 5 primeiras linhas
df.head(5)
```

```python
#Quantidade de Linhas e Colunas
df.shape
```

```python
#Verificando os tipos de dados
df.dtypes
```

```python
#Qual a Receita Total?
round(df['Valor Venda'].sum(),2)
```

```python
#Qual o Custo Total?
df['Custo'] = df['Custo Unitário'].mul(df['Quantidade'])
```

```python
df.sample(5)
```

```python
#Qual o custo total?
round(df['Custo'].sum(),2)
```

```python
#Coluna de Lucro Obtido
df['Lucro'] = df['Valor Venda'] - df['Custo']
```

```python
df.sample(5)
```

```python
#Lucro Total
round(df['Lucro'].sum(),2)
```

```python
#Tempo total de dias para enviar o produto
df['Tempo Envio'] = df['Data Envio'] - df['Data Venda']
```

```python
df.sample(5)
```

**Agora iremos analisar a média de tempo de envio para cada Marca**

```python
df['Tempo Envio'] = (df['Data Envio'] - df['Data Venda']).dt.days
```

```python
df.sample(1)
```

```python
#Média do tempo de envio por Marca
df.groupby('Marca')['Tempo Envio'].mean()
```

**Missing Values**

```python
#Verificando se existe valores nulos
df.isnull().sum()
```

**Lucro por ano e por marca**

```python
#Agrupamento por ano e marca
df.groupby([df['Data Venda'].dt.year, 'Marca'])['Lucro'].sum()
```

```python
pd.options.display.float_format = '{:20,.2f}'.format
```

```python
#Resetando o index
lucro_ano = df.groupby([df['Data Venda'].dt.year, 'Marca'])['Lucro'].sum().reset_index()
lucro_ano
```

```python
#Qual o total de produtos vendidos?
df.groupby('Produto')['Quantidade'].sum().sort_values(ascending=False)
```

```python
#Gráfico Total de Produtos Vendidos
df.groupby('Produto')['Quantidade'].sum().sort_values(ascending=True).plot.barh(title='Produtos Total Vendidos')
plt.xlabel('Total')
plt.ylabel('Produto');
```

```python
df.groupby(df['Data Venda'].dt.year)['Lucro'].sum().plot.bar(title='Lucro x Ano')
plt.xlabel('Ano')
plt.ylabel('Lucro');
```

```python
df.groupby(df['Data Venda'].dt.year)['Lucro'].sum()
```

```python
#Vendas de 2019
df_2009 = df[df['Data Venda'].dt.year == 2009]
```

```python
df_2009.sample(5)
```

```python
df_2009.groupby(df_2009['Data Venda'].dt.month)['Lucro'].sum().plot(title='Lucro x Mês em 2009')
plt.xlabel('Mês')
plt.ylabel('Lucro');
```

```python
df_2009.groupby('Marca')['Lucro'].sum().plot.bar(title='Lucro x Marca em 2009')
plt.xlabel('Marca')
plt.ylabel('Lucro')
plt.xticks(rotation='horizontal');
```

```python
df_2009.groupby('Classe')['Lucro'].sum().plot.bar('Lucro x Classe em 2009')
plt.xlabel('Classe')
plt.ylabel('Lucro')
plt.xticks(rotation='horizontal');
```
